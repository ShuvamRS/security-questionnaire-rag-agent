# Foundry Connection Test

This notebook checks whether the local environment can connect to the Azure AI Foundry project and list the project connections.

```python
from pathlib import Path
from dotenv import load_dotenv
import os

from azure.identity import DefaultAzureCredential
from azure.ai.projects import AIProjectClient

# This notebook is inside walkthrough/, so the .env file is one level above it.
load_dotenv(dotenv_path=Path("../.env"))

AIPROJECT_ENDPOINT = os.getenv("AIPROJECT_ENDPOINT")

# DefaultAzureCredential uses the Azure login from the local machine.
project_client = AIProjectClient(
    endpoint=AIPROJECT_ENDPOINT,
    credential=DefaultAzureCredential()
)

print("AIProjectClient created successfully.")
```

```python
# Check project connections.
connections = list(project_client.connections.list())

print("Project connections found:", len(connections))

search_connection_found = False

for connection in connections:
    if connection.type == "CognitiveSearch":
        search_connection_found = True

print("Azure AI Search connection found:", search_connection_found)
```
